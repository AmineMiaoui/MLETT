```python
from collections import deque

def minimal_operations(n, p, banned, k):
    answer = [-1] * n
    banned_set = set(banned)
    answer[p] = 0
    queue = deque([(p, 0)])

    while queue:
        current_pos, operations = queue.popleft()
        for start in range(max(0, current_pos - k + 1), min(n - k + 1, current_pos + 1)):
            if start + k > n:
                continue

            new_pos = start + (k - 1) - (current_pos - start)

            if new_pos in banned_set:
                continue

            if answer[new_pos] == -1:
                answer[new_pos] = operations + 1
                queue.append((new_pos, operations + 1))

    return answer
```

```python
n = 4
p = 0
banned = [1,2]
k = 4

result = minimal_operations(n, p, banned, k)
print("Résultat :", result)
```

```python
import random

n = random.randint(2, 10**5)
p = random.randint(0, n - 1)

banned_length = random.randint(0, n - 1)

banned = set()
while len(banned) < banned_length:
    val = random.randint(0, n - 1)
    if val != p:
        banned.add(val)
banned = list(banned)

k = random.randint(1, n)

print(f"n = {n}")
print(f"p = {p}")
print(f"banned = {banned}")
print(f"k = {k}")
```

```python
result = minimal_operations(n, p, banned, k)
print("Résultat :", result)
```
